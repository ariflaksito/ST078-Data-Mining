# Naive Bayes Classifier (NBC)

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('dataset_mhs.xlsx')
df
```

```python
number = LabelEncoder()
df['JK'] = number.fit_transform(df['JK'])
df['TEPAT WAKTU'] = number.fit_transform(df['TEPAT WAKTU'])
```

```python
df
```

```python
df_train = df[0:84]
```

```python
df_train
```

```python
df_test = df[84:]
```

```python
df_test
```

```python
X_train = df_train.iloc[:,1:4]
y_train = df_train.iloc[:,4:]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train
```

```python
y_train
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
X_test = df_test.iloc[:,1:4]
X_test
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
pred_proba = model.predict_proba(X_test)
```

```python
pred_proba
```
